```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data_2015 = pd.read_csv("C:/Users/SOUMYA K THAKAI/Downloads/2015.csv",)
data_2015["year"]=2015
data_2015.drop_duplicates(inplace=True)
data_2015.dropna()
data_2015.rename(columns={"Happiness Score":"Happiness_score","Health (Life Expectancy)":"Health","Economy (GDP per Capita)":"GDP"},inplace=True)
```

```python
plt.figure(figsize = (8,6))
ax = sns.barplot(x=data_2015.Region,y=data_2015.Happiness_score )
plt.xticks(rotation=15,ha='right')
plt.title("Happiness_socre v/s Region")
plt.show()
```

```python
sns.pairplot(data_2015)
plt.xticks(rotation = 90,ha='right')
plt.show()
```

```python
sns.regplot(data=data_2015,x=data_2015.Happiness_score,y=data_2015.Health)
plt.xlim(0,8)
plt.ylim(0.1,2)
plt.title("Health and Happiness Score relation")
plt.show()
```

```python
sns.regplot(data=data_2015,x=data_2015.GDP,y=data_2015.Happiness_score)
plt.title("GDP v/s Happiness Score relation")
plt.show()
```
